```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data= pd.read_csv("/content/cars-1.csv")
data.head()
```

```python
car_features= data[['mpg', 'disp', 'hp']].copy()
car_features.head()
```

```python
x= car_features.iloc[:,0:3]
kmeans= KMeans(3)
kmeans.fit(x)
individual_clusters = kmeans.fit_predict(x)

data_with_clusters = car_features.copy()
data_with_clusters['Clusters'] = individual_clusters 

plt.figure(figsize=[8,6])
plt.scatter(data_with_clusters['mpg'],data_with_clusters['disp'],
            c=data_with_clusters['Clusters'],cmap='rainbow')
plt.grid()
```

```python
x= car_features.iloc[:,0:3]
kmeans= KMeans(5)
kmeans.fit(x)
individual_clusters = kmeans.fit_predict(x)

data_with_clusters = car_features.copy()
data_with_clusters['Clusters'] = individual_clusters 

plt.figure(figsize=[8,6])
plt.scatter(data_with_clusters['mpg'],data_with_clusters['hp'],
            c=data_with_clusters['Clusters'],cmap='rainbow')
plt.grid()
```
