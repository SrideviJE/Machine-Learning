```python
import numpy as np
import pandas as pd
```

```python
timestamp= ['2020-02-16 16:38:56', '2020-08-15 09:18:46', '2020-01-06 02:58:16', '2020-06-06 06:38:56', '2020-06-06 06:38:56']
```

```python
data= {
    'timestamp': timestamp,
    'transient_1': [11,75,7,7,7],
    'transient_2': [735,85,7,96,1],
    'transient_3': [5,45,9,78,12],
}
```

```python
df= pd.DataFrame(data= data)
df.head()
```

```python
df.iloc[1:2, 1:4]= [1,2,3]
df
```

```python
df['Average']= df.mean(axis=1)
df
```

```python
df['month']= pd.DatetimeIndex(df['timestamp']).month
df
```

```python
dfs= df.iloc[:,1:4].to_numpy()
b= np.sort(dfs)
df['average_7_max']= pd.DataFrame(np.mean(b[:, 1:], axis=1))
df
```

```python
columns= ['transient_1','transient_2','transient_3']
for i, row in df.iterrows():
    for column in columns:
      if ((df.at[i,column]>= 50) and (df.at[i,column]<= 800)):
            df.at[i,column] = row[column] / 10
df
```

```python
df.sort_values(by= ['timestamp','transient_1','transient_2' ], inplace= True)
df
```
