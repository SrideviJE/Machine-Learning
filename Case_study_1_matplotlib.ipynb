```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cust= pd.read_csv("/content/customer_churn.csv")
```

```python
sns.countplot(x= "Contract", data= cust, color= 'orange')
plt.xlabel("Contract Type of customer")
plt.ylabel("Count")
plt.title("Distribution of Contract")
plt.show()
```

```python
cust["MonthlyCharges"].plot(kind= 'hist', color='forestgreen')
plt.xlabel("Monthly Charges Incurred")
plt.ylabel("Count")
plt.title("Distribution of Monthly Charges")
plt.show()
```

```python
cust.plot(kind= 'scatter', x= "tenure", y= "TotalCharges", color='indigo', s= 0.5)
plt.xlabel("Tenure of the customer")
plt.ylabel("Total charges Incurred")
plt.title("Total Charges vs Tenure")
plt.show()
```

```python
cust.boxplot(by= 'PaymentMethod', column= ['MonthlyCharges'], color='olive', grid=False)
plt.xlabel("Payment Method of customer")
plt.xticks(rotation=90)
plt.ylabel("Monthly Charges Incurred")
plt.title("Monthly Charges vs. Payment Method")
plt.show()
```
