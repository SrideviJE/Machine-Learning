```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data= pd.read_csv("/content/customers.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=[6,6])
plt.scatter(data["Age"], data["Spending Score (1-100)"], color="orange" )
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.grid()
plt.show()
```

```python
wcss=[]
x= data.iloc[:,2:4]
for i in range(1,11):
  kmeans= KMeans(i)
  kmeans.fit(x)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)
print(wcss)

number_clusters= np.arange(1,11)
print(number_clusters)
plt.figure(figsize=[6,6])
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid()
```

```python
x= data.iloc[:,2:4]
kmeans= KMeans(2)
kmeans.fit(x)
identified_clusters = kmeans.fit_predict(x)

data_with_clusters = data.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.figure(figsize=[6,6])
plt.scatter(data_with_clusters['Age'],data_with_clusters['Annual Income (k$)'],
            c=data_with_clusters['Clusters'],cmap='rainbow')
plt.grid()
```
