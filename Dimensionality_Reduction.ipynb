```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.decomposition import FactorAnalysis as FA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
digits= load_digits()
print(digits.data.shape)
target_names= digits.target_names
y= digits.target
```

```python
plt.figure(figsize=[6,6])
pca= PCA()
A= pca.fit_transform(digits.data)
print(A.shape)
for i in range(10):
   plt.scatter(A[y==i, 0], A[y==i,1], s=15)
plt.legend(labels= target_names)
plt.title('PCA of DIGITS dataset')
plt.show()
```

```python
plt.figure(figsize=[6,6])
fa= FA(n_components=2)
B= fa.fit_transform(digits.data, digits.target)
print(B.shape)
for i in range(10):
   plt.scatter(B[y==i, 0], B[y==i, 1] ,s= 15)
plt.legend(labels= target_names)
plt.title('FA of DIGITS dataset')
plt.show()
```

```python
plt.figure(figsize=[6,6])
lda= LDA(n_components=2)
X_train= lda.fit_transform(digits.data,digits.target)
print(X_train.shape)
for i in range(10):
  plt.scatter(X_train[y==i, 0], X_train[y==i,1], s=15)
plt.legend(labels= target_names)
plt.title('LDA of DIGITS dataset')
plt.show()
```
